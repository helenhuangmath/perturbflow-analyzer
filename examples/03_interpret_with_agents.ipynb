# 03. Interpret Results With Interpretation Handoff

`perturbflow interpret` summarizes derived result tables into Markdown and JSON files that can be reviewed by a scientist or used in a structured review workflow. It does not include raw count matrices.

````python
from pathlib import Path

from perturbflow.ai import collect_result_context, write_agent_handoff

PROJECT = Path.cwd()
RESULTS_DIR = PROJECT / "results" / "my_first_run"
PROJECT_NAME = "My Perturb-seq screen"

RUN_EXPORT = False
````

````python
if RUN_EXPORT:
    files = write_agent_handoff(
        results_dir=RESULTS_DIR,
        project_name=PROJECT_NAME,
        max_rows=15,
    )
    for label, path in files.items():
        print(f"{label}: {path}")
else:
    print("Set RUN_EXPORT=True after RESULTS_DIR points to a completed run.")
````

## Inspect context before sharing

Always review the generated files before sending them to an external model provider.

````python
context_path = RESULTS_DIR / "agent_handoff" / "interpretation_context.md"
if context_path.exists():
    print(context_path.read_text(encoding="utf-8")[:3000])
else:
    print(f"No context file found yet: {context_path}")
````

````python
if RESULTS_DIR.exists():
    context = collect_result_context(RESULTS_DIR, max_rows=5)
    print(context.keys())
    print("Top DEG rows:", len(context.get("deg_top", [])))
else:
    print("RESULTS_DIR does not exist yet.")
````

## Equivalent command line

```bash
perturbflow interpret \

  --project-name "My Perturb-seq screen" \
```
