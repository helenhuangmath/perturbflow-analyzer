# 04. Explore PerturbFlow Outputs

Use this notebook after a completed run to inspect summary JSON, key CSV tables, reports, plots, and the bundle manifest.

```python
import json
from pathlib import Path

import pandas as pd

PROJECT = Path.cwd()
RESULTS_DIR = PROJECT / "results" / "my_first_run"
```

```python
summary_path = RESULTS_DIR / "summary.json"
checkpoint_path = RESULTS_DIR / "checkpoint.json"

for path in [summary_path, checkpoint_path]:
    if path.exists():
        print(path.name)
        print(json.dumps(json.loads(path.read_text()), indent=2)[:2000])
    else:
        print(f"Missing: {path}")
```

## Differential expression summary

```python
deg_path = RESULTS_DIR / "csv" / "deg_summary.csv"
if deg_path.exists():
    deg = pd.read_csv(deg_path)
    display(deg.sort_values("n_de_total", ascending=False).head(10))
else:
    print(f"Missing: {deg_path}")
```

## Connectivity rewiring summary

```python
cscore_path = RESULTS_DIR / "csv" / "cscore_summary.csv"
if cscore_path.exists():
    cscore = pd.read_csv(cscore_path)
    display(cscore.sort_values("c_total", ascending=False).head(10))
else:
    print(f"Missing: {cscore_path}")
```

## Reports and bundle

```python
paths = [
    RESULTS_DIR / "interactive_report.html",
    RESULTS_DIR / "report.html",
    RESULTS_DIR / "bundle" / "manifest.json",
]

for path in paths:
    print(f"{path}: {'found' if path.exists() else 'missing'}")

manifest_path = RESULTS_DIR / "bundle" / "manifest.json"
if manifest_path.exists():
    manifest = json.loads(manifest_path.read_text())
    print(json.dumps(manifest, indent=2)[:2000])
```
