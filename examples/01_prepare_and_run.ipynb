# 01. Prepare Data And Run PerturbFlow

This notebook shows the standard workflow: install/import PerturbFlow, standardize a user `.h5ad`, run the analysis, and locate the main reports.

Update the paths in the setup cell before running on your dataset.

## Install

From a terminal in the repository root:

```bash
python -m pip install -e ".[bundle]"
```

On the Wherry cluster, activate the existing environment first if you already have one.

````python
from pathlib import Path

from perturbflow import prepare_h5ad, run_analysis

PROJECT = Path.cwd()
RAW_H5AD = Path("/path/to/my_raw_perturbseq_data.h5ad")
PREPARED_H5AD = PROJECT / "prepared" / "my_data.perturbflow.h5ad"
RESULTS_DIR = PROJECT / "results" / "my_first_run"
CONFIG = PROJECT / "configs" / "cluster_default.json"

PERTURBATION_COL = "guide_gene"
CONTROL_LABELS = "control,non-targeting,NT"
CELL_STATE_COL = "leiden"  # set to None if your input has no cell-state column

RUN_ANALYSIS = False  # Change to True after paths are correct.
````

````python
if RUN_ANALYSIS:
    prepared = prepare_h5ad(
        input_path=RAW_H5AD,
        output_path=PREPARED_H5AD,
        perturbation_col=PERTURBATION_COL,
        control_labels=CONTROL_LABELS,
        cell_state_col=CELL_STATE_COL,
    )
    print(f"Prepared AnnData written to: {prepared}")
else:
    print("Set RUN_ANALYSIS=True after updating RAW_H5AD and column names.")
````

````python
if RUN_ANALYSIS:
    run_analysis(
        input_path=PREPARED_H5AD,
        output_dir=RESULTS_DIR,
        config_path=CONFIG,
        resume=True,
    )
    print(f"Results written to: {RESULTS_DIR}")
else:
    print("Skipping pipeline run in template mode.")
````
